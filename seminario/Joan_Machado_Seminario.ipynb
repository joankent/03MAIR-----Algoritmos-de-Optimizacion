(*) Diseña un algoritmo que mejore la complejidad del algoritmo por fuerza bruta

```
# Buscamos el par de numeros en el array para poder buscar que la suma sea 7
# Y nos retorna las posiciones de esos pares

v = list(range(1,7,1))
v
```

Por fuerza bruta analizamos cada posibilidad en un recorrido de la lista

```
def searchPairSimple(data, target):
  for i in range(len(data)):
    for j in range(len(data)):
      if i == j:
        continue
      if (data[i] + data[j] == target):
        return [i,j]
```

Tardará aproximadamente 7.2479e-05 unidades de tiempo

```
from time import time

tiempo_inicial = time() 
searchPairSimple(v,7)
tiempo_final = time() 

tiempo_ejecucion = tiempo_final - tiempo_inicial
tiempo_ejecucion
```

Almacenamos los tiempos de una lista

```
import numpy as np

size = range(250,2750, 250)
tiempo = []
for i in size:
  v = list(np.linspace(1,7,i))
  tiempo_inicial = time() 
  searchPairSimple(v,7)
  tiempo_final = time() 
  tiempo_ejecucion = tiempo_final - tiempo_inicial
  tiempo.append(tiempo_ejecucion)
tiempo
```

Optimizamos nuestro algoritmo inicializando la j al valor siguiente en el segundo bucle

```
def searchPairOptimized(data, target):
  for i in range(len(data) - 1):
    for j in range(i + 1, len(data)):
      if (data[i] + data[j] == target):
        return [i,j]
```

```
import numpy as np

tiempo_ = []
for i in size:
  v = list(np.linspace(1,7,i))
  tiempo_inicial = time() 
  searchPairOptimized(v,7)
  tiempo_final = time() 
  tiempo_ejecucion = tiempo_final - tiempo_inicial
  tiempo_.append(tiempo_ejecucion)
```

Vemos una increible mejora en el tiempo de ejecución del algoritmo con esa modificación

```
import matplotlib.pyplot as plt

plt.plot(size, tiempo, label= 'searchPairSimple')
plt.plot(size, tiempo_,label= 'searchPairOptimized')
plt.grid()
plt.xlabel('Cantidad de datos')
plt.ylabel('Tiempo de ejecución')
plt.legend()
plt.show()
```
